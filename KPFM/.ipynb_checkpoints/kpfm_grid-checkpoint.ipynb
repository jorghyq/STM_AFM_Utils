```python
base_dir = './test/'
files = os.listdir(base_dir)
size = 30
grid = np.zeros((size,size))
for item in files:
    match = re.search(r'KPFM_3',item)
    if match:
        ind = int(item[7:10]) - 1
        #print ind
        ind_x = ind % size
        ind_y = ind / size
        coeff = fit_freq(base_dir + item)
        grid[ind_x,ind_y] = coeff[1]/(2*coeff[0])
```

```python
plt.imshow(grid,interpolation='nearest')
```
